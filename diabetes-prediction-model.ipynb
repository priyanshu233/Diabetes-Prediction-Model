```python
X = ks.drop(columns ='Outcome',axis=1)
Y = ks['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

```python
scaler = StandardScaler()
```

In this code block,we fit the data and transform the data to make better predictions from the model.

```python
st1 = scaler.fit_transform(X)
```

```python
X = st1
Y = ks['Outcome'] 
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

*Model Evaluation*

Accuracy Score

```python
X_train_predictions = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predictions,Y_train)
```

```python
print(training_data_accuracy)
```

 It's a pretty good accuracy 

```python
X_test_predictions = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predictions,Y_test)
```

```python
print(test_data_accuracy)
```

Making a Predictive System

```python
input_data =(5,166,72,19,175,25.8,0.587,51)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0] == 0):
    print('The process is non-diabetic')
else:
    print('The person is diabetic')
```
